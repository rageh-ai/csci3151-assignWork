# **1. Decision Tree**

a.) In this question, you will predicting Hepatitis Using Decision Trees  [Hepatitis](https://archive.ics.uci.edu/ml/machine-learning-databases/hepatitis/). Experiment with all available features in sklearn library and generate comprehensive report of your findings.

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.model_selection import train_test_split

tree = DecisionTreeClassifier()

hep = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/hepatitis/hepatitis.data")

cols = ["Class","Age","Sex","Steroid","Antivirals","Fatigue", "Malaise",
        "Anorexia", "Liver Big", "Liver Firm", "Spleen Palpable", "Spiders",
        "Ascites", "Varices", "Bilirubin", "ALK Phosphate", "SGOT", "Albumin",
        "Protime", "Histology"]

hep.columns = cols



X = hep.iloc[:,1:20]
X = X.drop("Protime", axis=1)
X = X.replace("?", 0)
y = hep.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25,\
                                                    random_state = 42)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)


acc = sum(y_pred == y_test)/len(y_pred)

print(acc)


```
