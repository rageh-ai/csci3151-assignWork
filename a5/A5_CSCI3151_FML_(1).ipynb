## **1. K-Means and Agglomerative Clustering**











In this question we are going to explore two different clustering methods on the [Wine Dataset](https://archive.ics.uci.edu/ml/datasets/Wine) and evaluate it using two measures: one is an intrinsic measure (no labels), while the other one makes use of the available labels.


```python
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import numpy as np
from sklearn.metrics import adjusted_rand_score
import pandas as pd
from sklearn.cluster import DBSCAN




X,y = datasets.load_wine(return_X_y=True)
```

**Hint :**



To evaluate an intrinsic measure, we can use metrics such as Silhouette score, Calinski-Harabasz index, or Davies-Bouldin index. These metrics evaluate the quality of the clustering results based on the internal structure of the data, without using any external information or labels. To evaluate using labels, we can use metrics such as Adjusted Rand Index (ARI), Normalized Mutual Information (NMI), or Fowlkes-Mallows index. These metrics compare the clustering results with the known labels to measure how well the algorithm is able to group the data into their true categories.

a) Cluster the dataset using the [Agglomerative Clustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html), and [k-Means](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans) clustering algorithm without using the class information as part of the features. Experiment with different numbers of clusters ranging from 2 to 5.



```python
sil_agg = []
ari_agg = []
sil_k_list = []
ari_k_list = []

for n in range(2,6):
  agg_cluster = AgglomerativeClustering(n_clusters = n)
  target = agg_cluster.fit_predict(X)

  k_cluster = KMeans(n_clusters = n)
  target_k = k_cluster.fit_predict(X)

  sil = silhouette_score(X, target)
  sil_k = silhouette_score(X,target_k)
  ari = adjusted_rand_score(y,target)
  ari_k = adjusted_rand_score(y,target_k)

  sil_agg.append(sil)
  sil_k_list.append(sil_k)
  ari_agg.append(ari)
  ari_k_list.append(ari_k)


ideal_agg = sil_agg.index(max(sil_agg)) + 2
ideal_k = sil_k_list.index(max(sil_k_list))+ 2
```

**`Hint:`** To find the optimal number of clusters in the 2-5 range, you can use the Elbow method or Silhouette score. The Elbow method involves plotting the Within-Cluster-Sum-of-Squares (WCSS) against the number of clusters and choosing the number of clusters at the elbow point, where the curve starts to flatten out. The Silhouette score measures the compactness and separation of the clusters and ranges from -1 to 1, with values closer to 1 indicating better clustering. We can choose the number of clusters that maximize the Silhouette score.

b) What is the variability of the resulting clusters as a function of different initializations or parameterization? Use the Silhouette coefficient and Adjusted Rand Index as metrics for evaluation to discuss the stability of results.

```python
print(f"The Silhouette Scores obtained by testing the Agglomerative clustering with 2-5 clusters have a variance of {np.var(sil_agg):.4f}")
print(f"The Silhouette Scores obtained by testing the K-means clustering with 2-5 have a variance of {np.var(sil_k_list):.4f}")


print(f"The Adjusted Rand scores obtained by testing Agglomerative clustering with 2-5 clusters have a variance of {np.var(ari_agg):.4f}")
print(f"The Adjusted Rand scores obtained by testing K-means with 2-5 clusters have a variance of {np.var(ari_k_list):.4f}")


```

**Hint:** Cluster variability refers to the degree of similarity or difference between clustering results obtained by running the algorithm with different parameter settings or initialization. Clustering metrics like Silhouette score and Adjusted Rand Index (ARI) can be used to measure the stability of results. A higher Silhouette score indicates better clustering while a higher ARI score indicates a better match with the true labels. We can compare the scores obtained with different parameter settings or initialization to evaluate the stability of the results.

## **2. DBSCAN Clustering**

In this question we are going to explore two different clustering methods on the [Wheat Seeds Dataset](https://archive.ics.uci.edu/ml/datasets/seeds) and evaluate it using two measures: one is an intrinsic measure (no labels), while the other one makes use of the available labels.


```python
dataset = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt",sep='\s+', header=None, dtype=np.float64)
dataset.head()

y = dataset.iloc[:,7]
dataset = dataset.drop(labels=7,axis=1)
```

a) Cluster the dataset using DBSCAN without using the class information as part of the features. Experiment with different values for the parameters Eps and minPts. It’s up to you how you deal with any outliers, but make sure your criterion is properly justified. What is the variability of the
resulting clusters as a function of different initializations? Use the Silhouette coefficient and Adjusted Rand Index as metrics for evaluation.

```python
params = np.arange(0.5,2,0.5)
sil_scores = []

for x in params:
  cluster = DBSCAN(eps=x)
  target = cluster.fit_predict(dataset)
  sil_score = silhouette_score(dataset,target)
  sil_scores.append(sil_score)


print(f"The variance of the silhouette scores is: {np.var(sil_scores):.2f}")
print(f"The highest silhouette score was: {max(sil_scores):.2f} at eps = {params[sil_scores.index(max(sil_scores))]}")
```
