b) We want to train a logistic regression model to predict the Fraud detection . It’s important that no other external package is used (pandas, numpy are ok) **for this question part** . We want to find the weights for the logistic regression using a hand made gradient descent algorithm. We will use cross-entropy as the cost function, and the logistic cross-entropy to compute the weight update during gradient descent. We are now using a random 20% of your data instances for validation.

Your function should be able to return the updated weights and bias after every iteration of the gradient descent algorithm.

Your function should be defined as follows:

``
def LRGradDesc( data, target, weight_init, bias_init, learning_rate, max_iter ):``

And it should print lines as indicated below (note the last line with the weights):

`Iteration 0: [initial_train cost], [train accuracy], [validation accuracy]`

`Iteration 1: [train cost after first iteration], [train accuracy after first iteration], [validation accuracy after first iteration]`

`Iteration 2: [weights after second iteration], [train cost after second iteration], [train accuracy after second iteration], [validation accuracy after second iteration]`

…

`Iteration max_iter : [weights after max_iter iteration], [train cost after max_iter iteration], [train accuracy after max_iter iteration], [validation accuracy after max_iter iterations]`

`Final weights: [bias], [w_0], [w_1]`

Note that you may want to print every 100 or every 1000 iterations if `max_iter` is a fairly large number (but you shouldn’t have more iterations than the indicated in `max_iter`).

````python
#Question B:

def LRGradDesc( data, target, weight_init, bias_init, learning_rate, max_iter, sample_size):

  #initializing bias
  b = np.full(shape=(len(data),1), fill_value=bias_init)
  #appending bias
  X_b = np.append(data,b,axis=1)
  #initializing theta
  theta = np.full(shape=(X_b.shape[1],1),fill_value= weight_init)


  X_train, X_test, y_train, y_test = train_test_split(X_b,target,random_state=42,test_size=0.2)
  y_test = y_test.to_numpy().reshape(-1,1)

  np.seterr(divide="ignore",over='ignore')


  for iter in range(max_iter):

    #sampling for computational efficiency
    index = np.random.choice(X_train.shape[0], size=sample_size, replace=False)
    sample_X = X_train[index]
    sample_y = np.ravel(y_train.iloc[index]).reshape(-1,1)




    #computing predictions using sigmoid function
    z = np.dot(sample_X, theta)
    z_test = np.dot(X_test, theta)
    y_pred = np.ravel((1/(1 + np.exp(-z)))).reshape(-1,1)
    y_tpred = np.ravel((1/(1+np.exp(-z_test)))).reshape(-1,1)


   #computing loss, used a function here as it kept crashing when I implemented by own cross entropy loss
    log_pred = np.log(y_pred)*sample_y
    loss = np.sum(np.multiply(sample_y,log_pred) + np.multiply(1-sample_y, log_pred))



    #computing gradient
    diff = (y_pred - sample_y)

    grad = 1/sample_size * np.dot(sample_X.T,diff)

    #optimization
    theta  = theta - (learning_rate*grad)


    #classifictaion
    y_pred = np.rint(y_pred)
    y_tpred = np.rint(y_tpred)

    #calculating acc
    sample_acc = accuracy_score(sample_y,y_pred)
    dev_acc = accuracy_score(y_test,y_tpred)



    print(f"Iteration {iter}:")
    print(f"sample accuracy: {sample_acc}")
    print(f"dev accuracy: {dev_acc}")
    print(f"weights: {np.ravel(theta)}")
    print("--------------------")






````
